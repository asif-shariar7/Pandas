<h2 align="center">Learning Pandas</h2>

### What is Pandas?
<small>Pandas is a Python library for <b>data analysis</b> and <b>manipulation</b>.  
It makes working with tables of data (like Excel sheets or CSV files) very easy.  
Think of it as a tool that helps us clean, explore, and analyze data quickly.</small>

### Main Features :
<small>
<b>Handles tabular data</b> – Rows and columns (like a spreadsheet).<br> 
<b>Data cleaning</b> – Fill missing values, remove duplicates, fix formats.<br>  
<b>Data selection</b> – Easily pick certain rows, columns, or filter data.<br> 
<b>Merging & joining</b> – Combine multiple datasets.<br>
<b>Grouping & aggregation</b> – Summarize data (like average, sum, count).<br>
<b>Time series support</b> – Work with dates and times efficiently.<br>
<b>Note:</b> Pandas are also able to delete rows that are not relevant, or contains wrong values, like empty or NULL values. This is called cleaning the data.<br>
</small>

<p>A Pandas <b>Series</b> is like a column in a table. It is a one-dimensional array holding data of any type.</p>

````python
arr = [8,9,0]
myvar = pd.Series(arr)
print(myvar)
print(myvar[1])
````

<p>Data sets in Pandas are usually multi-dimensional tables, called DataFrames. Series is like a column, a DataFrame is the whole table.</p>

````python
data = {'Name' : ['A','B','C'],
        'Age': [22,11,24]}

df = pd.DataFrame(data)
print(df)

# with index argument
df_indexed = pd.DataFrame(data, index = ['x','y','z'])
print(df_indexed)
````

<p><b>drop()</b> in Pandas - </p>
<p>Syntax : DataFrame.drop(labels=None, axis=0, index=None, columns=None, inplace=False)</p>
<p>Here, labels → Name(s) of row/column to remove. <br>
axis 0 → drop rows and 1 → drop columns. <br>
index → specify row labels directly. <br>
columns → specify column names directly. <br>
inplace = True → modify original DataFrame. False → return new DataFrame (default). </p>

````python
data = {
    "Name": ["A", "B", "C"],
    "Age": [20, 21, 22],
    "City": ["Dhaka", "Chittagong", "Khulna"]
}

df = pd.DataFrame(data)
print(df)
````

<p>Drop a row </p>

````python
df.drop(1, axis = 0)
````

<p>Drop Multiple Rows</p>

````python
df.drop([0,2], axis= 0)
````

<p>Drop a column</p>

````python
df.drop('Age', axis = 1)
````

<p>Drop multiple columns</p>

````python
df.drop(columns=['Age', 'City'])
````

<p>Drop Using <b>index</b> Parameter</p>

````python
df.drop(index=1)
````

<p>In Pandas, the <b>drop()</b> method does not modify the original DataFrame by default; it returns a new DataFrame with the specified rows or columns removed.
To permanently delete data, we must use <b>inplace=True</b> or reassign the result to the original variable.</p>

````python
df
````

````python
df.drop(index=2, inplace=True)
````

````python
df
````

<p>Or</p>

````python
delete_permanent = df.drop(index=1)
delete_permanent
````

<p><b>loc</b> (Label-based indexing) : Selects data using row/column names (labels).</p>

````python
data = {"Name" : ["Asif", "Shariar", "Tashin"],
        "Age": [23,24,22]
        }

df = pd.DataFrame(data, index=['a','b','c'])
print(df)

print('\n')
print(df.loc['c'])
print('\n')
print(df.loc['b', 'Age'])
````

<p><b>iloc</b> (Position-based indexing) : Selects data using index position (0,1,2,...)</p>

````python
print(df.iloc[0])
print(df.iloc[1,1])  # row 1, column 1
# Get first two rows and first one columns
df.iloc[0:2, 0:1]
````

````python
print(df.iloc[[0,1],[0]])
print('\n')
print(df.loc[['a','c'], ['Name', 'Age']])
print("\n")
print(df[["Name", "Age"]])
````

<p>In Pandas, joining means combining multiple DataFrames into one.<br>
1. <b>merge()</b> → Database-style join<br>
Syntax : pd.merge(df1, df2, on="column_name", how="type")</p>

````python
df1 = pd.DataFrame({
    "ID": [1, 2, 3],
    "Name": ["Asif", "Shariar", "Tashin"]
})

df2 = pd.DataFrame({
    "ID": [1, 2, 4],
    "Age": [22, 23, 25]
})

merge_df = pd.merge(df1, df2, on = "ID", how="inner")
merge_df
````

<p>2. <b>concat()</b> → Stacking data</p>

````python
# Row-wise (default)
pd.concat([df1, df2])
````

````python
# Column-wise
pd.concat([df1, df2], axis=1)
````

<h3 align="center">Cleaning Data </h3>
<p>Data cleaning means fixing bad data in your data set.<br>
Bad data could be: 1. Empty cells 2. Data in wrong format 3. Wrong data 4. Duplicates</p>

<p>1. Cleaning Empty Cells - <br>
Empty cells can potentially give us a wrong result when we analyze data.<br>

1. <b>Remove Rows</b> <br>
One way to deal with empty cells is to remove rows that contain empty cells. This is usually OK, since data sets can be very big, and removing a few rows will not have a big impact on the result.<br>

2. <b>Replace Empty Values</b> <br>
Another way of dealing with empty cells is to insert a new value instead. This way you do not have to delete entire rows just because of some empty cells.<br>

3. <b>Replace Using Mean, Median, or Mode</b><br>
A common way to replace empty cells, is to calculate the mean, median or mode value of the column.</p>
