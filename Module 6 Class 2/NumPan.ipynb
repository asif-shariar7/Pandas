```python
import numpy as np
import pandas as pd
```

```python
data = {
    "Name" : ["Asif", "Nisa", "Shariar", np.nan],
    "Age" : [23, 21, np.nan, 24],
    "Salary" : [50000, 30000, 30000, np.nan]
}

df = pd.DataFrame(data)
```

```python
df
```

```python
df.fillna({
    "Name" : "Unknown",
    "Age" : df["Age"].mean(),
    "Salary" : 0
}, inplace = True)
```

```python
df
```

```python
df.dtypes
```

```python
df["Age"] = df["Age"].astype(int)
```

```python
df.dtypes
```

```python
df
```

```python
sorted_table = df.sort_values(by = "Salary", ascending = True)  # ascending = False for descending
sorted_table
```

<p><b>np.searchsorted(sorted_array, value)</b> finds the index where a value should be inserted in a sorted array to keep it sorted</p>

```python
arr = np.array([20,30,40,90])  
pos = np.searchsorted(arr, 15)
pos

# Note : Array must be sorted
```

<p><b>np.where(...)</b> - Returns the positions (indices) where condition is True. <b>np.where</b> returns a tuple</p>

```python
arr = np.array([201, 198, 77, 76, 29, 3])
even = np.where (arr % 2 == 0)
even 
```
