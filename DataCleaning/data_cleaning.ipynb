```python
import pandas as pd
```

```python
df = pd.read_csv("raw_data.csv")
```

```python
df.shape
```

```python
df.duplicated().sum()
```

```python
df.columns
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
y = df["Embarked"].mode()[0]
df.fillna({"Embarked" : y}, inplace=True)
```

```python
df.shape
```

```python
x = df["Age"].mean()
df.fillna({"Age" : x}, inplace=True)
```

```python
df.isnull().sum()
```

```python
df.drop("Cabin", axis=1, inplace=True)
```

<p><b>Note : </b> For delete a column - <b>df.drop("col_name", axis = 1)</b></p>

```python
df.isnull().sum()
```

```python
df.dtypes
```

<p>Fix Data Types</p>

```python
df["Age"] = df["Age"].astype(int)
```

```python
df.dtypes
```

<p>Handle Categorical Data <br>
Convert text → numeric</p>

```python
df["Sex"] = df["Sex"].map({'male' : 0, 'female': 1})
df["Embarked"] = df["Embarked"].map({"S": 0, "C": 1, "Q": 2})
```

```python
df.head()
```

<p>Drop Irrelevant Columns. These don’t help much in analysis</p>

```python
df.drop(["PassengerId", "Name", "Ticket"], axis=1, inplace=True)
```

```python
df.columns
```

```python
df.shape
```

```python
df.duplicated().sum()
```

```python
df[df.duplicated()]
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.describe()
```

<p>Last Checking</p>

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df.to_csv("clean_data.csv", index=False)
```
